```python
import pandas as pd
```

```python
descriptions = pd.read_csv('../data/descriptions.csv')
precautions = pd.read_csv('../data/precautions.csv')
treatments = pd.read_csv('../data/treatments.csv')
```

```python
merged = (descriptions.merge(precautions, on='Disease', how='outer')
          .merge(treatments, on='Disease', how='outer')
          .set_index('Disease'))
merged
```

```python
merged.to_csv('../data/diseases.csv')
```
